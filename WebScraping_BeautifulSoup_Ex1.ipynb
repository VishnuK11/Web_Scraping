```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
html_file = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Science').text 
output_dir = "./jobs"
if not os.path.isdir(output_dir):
    os.mkdir(output_dir)
```

```python
# Declare the parser for the file
soup = BeautifulSoup(html_file,'lxml')

# Find all job listings
jobs = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")

for index,job in enumerate(jobs):
    job_company = job.find("h3" ,class_="joblist-comp-name").text.replace('(More Jobs)','')
    job_company = job_company.replace("\r","").replace("\n","").replace("  ","")
    job_description = job.find("ul", class_="list-job-dtl clearfix").li.text.replace("\n","")
    job_date = job.find("span", class_="sim-posted").text.replace("\r","").replace("\n","")
    if "few days ago" in job_date:
        print(job_company)
        print(job_description)
        print(job_date)
        print()
        
        with open(f"{output_dir}/{index}.txt",'w') as f:
            f.write(job_company)
            f.write("\n")
            f.write(job_description)
            f.write("\n")
            f.write(job_date)
```
